```python
source_files_folder = "data/terramind_s1_for_submission_test/terramind_test_s1_emb"
source_predictions_folder = "runs/test_terramind_s1_decoder_residual/predictions"
output_predictions_folder_v1 = "runs/test_terramind_s1_decoder_residual/predictions_renamed_just_label_added"
output_predictions_folder_v2 = "runs/test_terramind_s1_decoder_residual/predictions_renamed_with_label_and_year"
output_predictions_folder_v3 = "runs/test_terramind_s1_decoder_residual/predictions_renamed_correctly"
```

```python
import os
import shutil

all_filenames = [filename for filename in os.listdir(source_files_folder)]
len(all_filenames), all_filenames[:2]
```

```python
all_predictions = [filename for filename in os.listdir(source_predictions_folder)]
len(all_predictions), all_predictions[:2]
```

```python
n = 1000
output_filenames_v1 = [
    'label_' + pred_name[5:12] + '.npy'
    for pred_name in all_predictions[:n]
]
len(output_filenames_v1), output_filenames_v1[:10]
```

```python
n = 1000
output_filenames_v2 = [
    'label_' + pred_name[5:12] + source_name[10:15] + '.npy'
    for pred_name, source_name in zip(all_predictions[:n], all_filenames[:n])
    if pred_name[5:12] == source_name[3:10]
]
len(output_filenames_v2), output_filenames_v2[:10]
```

```python
# given a list of files all_predictions from source_predictions_folder
# copy and rename them to output_predictions_folder_v1 with names from output_filenames_v1
os.makedirs(output_predictions_folder_v1, exist_ok=True)
for pred_name, new_name in zip(all_predictions[:n], output_filenames_v1):
    src_path = os.path.join(source_predictions_folder, pred_name)
    dst_path = os.path.join(output_predictions_folder_v1, new_name)
    shutil.copy2(src_path, dst_path)
```

```python
# given a list of files all_predictions from source_predictions_folder
# copy and rename them to output_predictions_folder_v2 with names from output_filenames_v2
os.makedirs(output_predictions_folder_v2, exist_ok=True)
for pred_name, new_name in zip(all_predictions[:n], output_filenames_v2):
    src_path = os.path.join(source_predictions_folder, pred_name)
    dst_path = os.path.join(output_predictions_folder_v2, new_name)
    shutil.copy2(src_path, dst_path)
```
